```python
class ParkingLot:
    def __init__(self):
        self.slots = 5

    def entry(self):
        if self.slots > 0:
            self.slots -= 1
            print("Vehicle parked")
        else:
            print("Parking full")

    def exit(self):
        hours = int(input("Enter parking hours: "))
        fee = hours * 20
        self.slots += 1
        print("Vehicle exited")
        print("Parking fee:", fee)

    def available(self):
        print("Available slots:", self.slots)


p = ParkingLot()

while True:
    print("\n1.Entry 2.Exit 3.Available Slots 4.Exit Program")
    choice = input("Enter choice: ")

    if choice == "1":
        p.entry()
    elif choice == "2":
        p.exit()
    elif choice == "3":
        p.available()
    elif choice == "4":
        break
```

```python
class ToDoList:
  def __init__(self):
    self.tasks = []

  def add_task(self):
    task = {
            "name": input("Task name: "),
            "deadline": input("Deadline: "),
            "priority": input("Priority : "),
            "category": input("Category: ")
        }
    self.tasks.append(task)
    print("Task added")

  def view_tasks(self):
    for t in self.tasks:
      print(t)

  def filter_by_category(self):
    cat = input("Enter category: ")
    for t in self.tasks:
      if t["category"] == cat:
        print(t)

x=ToDoList()
x.add_task()
x.view_tasks()
x.filter_by_category()
```

```python
class Billing_System:
    def __init__(self):
        self.products = {
            "chocolate": 10,
            "bread": 50,
            "icecream": 300
        }
        self.total = 0

    def scan_product(self):
        name = input("Enter product name: ")
        if name in self.products:
            self.total += self.products[name]
            print(name, "added")
        else:
            print("Product not found")

    def apply_discount(self):
        discount = int(input("Enter discount %: "))
        self.total -= self.total * discount / 100

    def generate_bill(self):
        print("Total Amount: ₹", self.total)
b=Billing_System()
b.scan_product()
b.scan_product()
b.apply_discount()
b.generate_bill()
```

```python
seats = 20

while True:
    print("\nAvailable seats:", seats)
    print("1.Book Ticket")
    print("2.Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        num = int(input("Number of tickets: "))

        if num <= seats:
            seats -= num
            print("Tickets booked successfully")
            print("Remaining seats:", seats)
        else:
            print("Not enough seats")

    elif choice == "2":
        break
```

```python
students = {}

name = input("Enter student name: ")

subjects = int(input("Number of subjects: "))
marks = []

for i in range(subjects):
    m = int(input("Enter marks: "))
    marks.append(m)

total = sum(marks)
average = total / subjects
gpa = average / 10

students[name] = {
    "marks": marks,
    "gpa": gpa
}

print("\nStudent Result")
print("Name:", name)
print("Marks:", marks)
print("GPA:", gpa)
```

```python
freelancers = {}
clients = {}
projects = {}

while True:
    print("\n1. Register Freelancer")
    print("2. Register Client")
    print("3. Assign Project")
    print("4. Process Payment")
    print("5. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        name = input("Enter freelancer name: ")
        skill = input("Enter skill: ")
        freelancers[name] = skill
        print("Freelancer registered.")

    elif choice == "2":
        name = input("Enter client name: ")
        clients[name] = True
        print("Client registered.")

    elif choice == "3":
        project = input("Enter project name: ")
        client = input("Enter client name: ")
        freelancer = input("Enter freelancer name: ")

        if client in clients and freelancer in freelancers:
            projects[project] = freelancer
            print("Project assigned to", freelancer)
        else:
            print("Client or Freelancer not found.")

    elif choice == "4":
        project = input("Enter project name: ")
        amount = input("Enter payment amount: ")

        if project in projects:
            freelancer = projects[project]
            print("Payment of", amount, "processed to", freelancer)
        else:
            print("Project not found.")

    elif choice == "5":
        break

    else:
        print("Invalid choice")
```

```python
inventory = {}

while True:
    print("\n1. Add Goods")
    print("2. Remove Goods")
    print("3. Inventory Report")
    print("4. Forecast Demand")
    print("5. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        item = input("Enter item name: ")
        qty = int(input("Enter quantity: "))
        inventory[item] = inventory.get(item, 0) + qty
        print("Goods added.")

    elif choice == "2":
        item = input("Enter item name: ")
        qty = int(input("Enter quantity to remove: "))
        if item in inventory and inventory[item] >= qty:
            inventory[item] -= qty
            print("Goods removed.")
        else:
            print("Not enough stock.")

    elif choice == "3":
        print("\nInventory Report")
        for item, qty in inventory.items():
            print(item, ":", qty)

    elif choice == "4":
        item = input("Enter item name: ")
        if item in inventory:
            forecast = inventory[item] + 10
            print("Forecast demand for", item, ":", forecast)
        else:
            print("Item not found.")

    elif choice == "5":
        break

    else:
        print("Invalid choice")
```
