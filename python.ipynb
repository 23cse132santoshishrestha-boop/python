```python
PIN = 1234
attempts = 3

while attempts > 0:
    entered_pin = int(input("Enter your PIN: "))

    if entered_pin == PIN:
        print("PIN verified successfully.\n")
        break
    else:
        attempts -= 1
        print(f"Wrong PIN. Attempts left: {attempts}")

if attempts == 0:
    print("Too many wrong attempts. Access denied.")
    exit()

bal = 0
trans = []

while True:
    print("\n1. Deposit")
    print("2. Withdraw")
    print("3. Check Balance")
    print("4. Transaction History")
    print("5. Exit")

    choice = int(input("Enter your choice = "))

    if choice == 1:
        amt = int(input("Enter the amount you want to deposit = "))
        bal += amt
        trans.append(f"Deposit: {amt}")
        print("Amount deposited successfully.")

    elif choice == 2:
        amt = int(input("Enter the amount to withdraw = "))
        if amt <= bal:
            bal -= amt
            trans.append(f"Withdraw: {amt}")
            print("Amount withdrawn successfully.")
        else:
            print("Insufficient balance.")

    elif choice == 3:
        print("Current Balance =", bal)

    elif choice == 4:
        if not trans:
            print("No transactions yet.")
        else:
            print("Transaction History:")
            for t in trans:
                print(t)

    elif choice == 5:
        print("Thank you!")
        break

    else:
        print("Invalid choice. Please try again.")
```

```python
PIN = 4444

def check_pin():
    attempts = 3
    while attempts > 0:
        pin = int(input("Enter your PIN: "))
        if pin == PIN:
            print("PIN verified successfully.\n")
            return True
        else:
            attempts -= 1
            print(f"Wrong PIN. Attempts left: {attempts}")
    print("Too many wrong attempts. Access denied.")
    return False


def deposit(bal, trans):
    amt = int(input("Enter the amount you want to deposit = "))
    bal += amt
    trans.append(f"Deposit: {amt}")
    print("Amount deposited successfully.")
    return bal


def withdraw(bal, trans):
    amt = int(input("Enter the amount to withdraw = "))
    if amt <= bal:
        bal -= amt
        trans.append(f"Withdraw: {amt}")
        print("Amount withdrawn successfully.")
    else:
        print("Insufficient balance.")
    return bal


def check_balance(bal):
    print("Current Balance =", bal)


def transaction_history(trans):
    if not trans:
        print("No transactions yet.")
    else:
        print("Transaction History:")
        for t in trans:
            print(t)


def atm_menu():
    bal = 0
    trans = []

    while True:
        print("\n1. Deposit")
        print("2. Withdraw")
        print("3. Check Balance")
        print("4. Transaction History")
        print("5. Exit")

        choice = int(input("Enter your choice = "))

        if choice == 1:
            bal = deposit(bal, trans)

        elif choice == 2:
            bal = withdraw(bal, trans)

        elif choice == 3:
            check_balance(bal)

        elif choice == 4:
            transaction_history(trans)

        elif choice == 5:
            print("Thank you!")
            break

        else:
            print("Invalid choice. Please try again.")


if check_pin():
    atm_menu()
```

10. Find repeated elements in a list using set


```python
lst = [1, 2, 3, 2, 4, 5, 1]

seen = set()
repeated = set()

for i in lst:
    if i in seen:
        repeated.add(i)
    else:
        seen.add(i)

print(repeated)
```

EVENT TRACKER

```python
entered_persons = set()
rejected_persons = set()

num_persons = 5

print("Event Tracker: Enter 5 persons\n")

for i in range(num_persons):
    name = input(f"Enter name of person {i+1}: ").strip()

    if name in entered_persons:
        print(f"Entry rejected! {name} has already entered.")
        rejected_persons.add(name)
    else:
        print(f"Entry accepted! Welcome {name}.")
        entered_persons.add(name)

print("\n--- Event Entry Summary ---")
print(f"Persons entered successfully ({len(entered_persons)}): {', '.join(entered_persons)}")
print(f"Persons rejected ({len(rejected_persons)}): {', '.join(rejected_persons)}")
print(f"Total persons who entered: {len(entered_persons)}")
print(f"Total persons rejected / duplicate entries: {len(rejected_persons)}")
print(f"Total persons not entered: {num_persons - len(entered_persons)}")
```

```python
entered_persons = set()
rejected_persons = set()

n = int(input("Enter the number of persons: "))

for i in range(n):
    name = input("Enter name of person: ")

    if name in entered_persons:
        print(name, "Entry rejected!")
        rejected_persons.add(name)
    else:
        print(name, "Entry accepted!")
        entered_persons.add(name)

print("Event Tracker!")

print("Total entered persons:")
for name in entered_persons:
    print(name)

print("Rejected persons:")
for name in rejected_persons:
    print(name)
```

```python
rooms = {
    101: None,
    102: None,
    103: None,
    104: None,
    105: None
}

while True:
    print("\n1. View Rooms")
    print("2. Book Room")
    print("3. Cancel Booking")
    print("4. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        for room, name in rooms.items():
            if name is None:
                print(room, "is available")
            else:
                print(room, "is booked by", name)

    elif choice == 2:
        room = int(input("Enter room number: "))
        if rooms.get(room) is None:
            name = input("Enter your name: ")
            rooms[room] = name
            print("Room booked successfully")
        else:
            print("Room already booked")

    elif choice == 3:
        room = int(input("Enter room number: "))
        if rooms.get(room) is not None:
            rooms[room] = None
            print("Booking cancelled")
        else:
            print("Room is already empty")

    elif choice == 4:
        print("Thank you!")
        break

    else:
        print("Invalid choice")
```

```python
students = {}   # student_id : {name, courses}
courses = {}    # course_id : course_name



def add_student():
    sid = input("Enter student ID: ")
    name = input("Enter student name: ")
    students[sid] = {"name": name, "courses": {}}
    print("Student added successfully!")



def add_course():
    cid = input("Enter course ID: ")
    cname = input("Enter course name: ")
    courses[cid] = cname
    print("Course added successfully!")


# Enroll student in course
def enroll_student():
    sid = input("Enter student ID: ")
    cid = input("Enter course ID: ")

    if sid in students and cid in courses:
        students[sid]["courses"][cid] = {"attendance": 0, "grade": None}
        print("Student enrolled successfully!")
    else:
        print("Invalid student or course ID")


# Mark attendance
def mark_attendance():
    sid = input("Student ID: ")
    cid = input("Course ID: ")

    if sid in students and cid in students[sid]["courses"]:
        students[sid]["courses"][cid]["attendance"] += 1
        print("Attendance marked!")
    else:
        print("Enrollment not found")



def assign_grade():
    sid = input("Student ID: ")
    cid = input("Course ID: ")
    grade = input("Enter grade: ")

    if sid in students and cid in students[sid]["courses"]:
        students[sid]["courses"][cid]["grade"] = grade
        print("Grade assigned!")
    else:
        print("Enrollment not found")


# Drop course
def drop_course():
    sid = input("Student ID: ")
    cid = input("Course ID: ")

    if sid in students and cid in students[sid]["courses"]:
        del students[sid]["courses"][cid]
        print("Course dropped successfully!")
    else:
        print("Enrollment not found")


# Student report
def student_report():
    sid = input("Student ID: ")

    if sid in students:
        print("\nStudent Name:", students[sid]["name"])
        for cid, details in students[sid]["courses"].items():
            print(f"Course: {courses[cid]}, Attendance: {details['attendance']}, Grade: {details['grade']}")
    else:
        print("Student not found")


# Course-wise report
def course_report():
    cid = input("Course ID: ")

    if cid in courses:
        print("\nCourse:", courses[cid])
        for sid, sdata in students.items():
            if cid in sdata["courses"]:
                details = sdata["courses"][cid]
                print(f"Student: {sdata['name']}, Attendance: {details['attendance']}, Grade: {details['grade']}")
    else:
        print("Course not found")


# Menu-driven program
while True:
    print("\n1.Add Student\n2.Add Course\n3.Enroll Student\n4.Mark Attendance\n5.Assign Grade")
    print("6.Drop Course\n7.Student Report\n8.Course Report\n9.Exit")

    choice = input("Enter choice: ")

    if choice == '1':
        add_student()
    elif choice == '2':
        add_course()
    elif choice == '3':
        enroll_student()
    elif choice == '4':
        mark_attendance()
    elif choice == '5':
        assign_grade()
    elif choice == '6':
        drop_course()
    elif choice == '7':
        student_report()
    elif choice == '8':
        course_report()
    elif choice == '9':
        print("Program ended.")
        break
    else:
        print("Invalid choice")
```

```python
books = []

def add_book():
    book_id = input("Enter Book ID: ")
    name = input("Enter Book Name: ")
    author = input("Enter Author Name: ")
    status = "Available"

    book = [book_id, name, author, status]
    books.append(book)
    print("Book added successfully")

def display_books():
    if len(books) == 0:
        print("No books available")
    else:
        for book in books:
            print("Book ID:", book[0])
            print("Book Name:", book[1])
            print("Author:", book[2])
            print("Status:", book[3])
            print("---------------------")

def issue_book():
    book_id = input("Enter Book ID to issue: ")
    found = False

    for book in books:
        if book[0] == book_id:
            found = True
            if book[3] == "Available":
                book[3] = "Issued"
                print("Book issued successfully")
            else:
                print("Book already issued")

    if found == False:
        print("Book not found")

def return_book():
    book_id = input("Enter Book ID to return: ")
    found = False

    for book in books:
        if book[0] == book_id:
            found = True
            if book[3] == "Issued":
                book[3] = "Available"
                print("Book returned successfully")
            else:
                print("Book was not issued")

    if found == False:
        print("Book not found")

while True:
    print("\nLibrary Management System")
    print("1. Add Book")
    print("2. Display Books")
    print("3. Issue Book")
    print("4. Return Book")
    print("5. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        add_book()
    elif choice == "2":
        display_books()
    elif choice == "3":
        issue_book()
    elif choice == "4":
        return_book()
    elif choice == "5":
        print("Thank you")
        break
    else:
        print("Invalid choice")
```
